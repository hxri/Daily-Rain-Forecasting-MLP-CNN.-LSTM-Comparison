# Data Analysis

The given dataset was on daily rainfall, from 1966-07-13 to 2021-04-14 (20000 days of rainfall data). The data is rea from the csv file and 2 plots, Line plot and a histogram plot is evaluated inorder to understand the trend and the shape of data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Timeseries_real_data_interview.csv")
values = data['Value'].values
```

### Line Plot and Histogram of unprocessed data

```python
plt.figure(figsize=(10,4))
plt.suptitle('Unprocessed Data', fontsize=16)
plt.subplot(1, 2, 1)
plt.plot(values)
plt.title('Line Plot')
plt.xlabel('Days')
plt.ylabel('Rainfall')

plt.subplot(1, 2, 2)
plt.hist(values)
plt.title('Histogram')

# spce between the plots
plt.tight_layout(3)
plt.savefig('data_analysis/raw_data.png')
# show plot
plt.show()
```

### Applying Box-Cox Transform on the raw data

```python
from scipy.stats import boxcox
transformed, lm = boxcox(values)
print('Statistically tuned Lambda Value: %s'%lm)
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(top=0.8)
plt.suptitle('Box-Cox Transformd Data (Lambda = 0.71035)', fontsize=16)
plt.subplot(1, 2, 1)
plt.plot(transformed)
plt.title('Line Plot')
plt.xlabel('Days')
plt.ylabel('Rainfall')

plt.subplot(1, 2, 2)
plt.hist(transformed)
plt.title('Histogram')

# spce between the plots
plt.tight_layout(3)
plt.savefig('data_analysis/transformed_data.png')
# show plot
plt.show()
```
